### Collaborative Filtering

```python
#read in frequency table.
frequency = pd.read_csv('frequency.csv')
frequency.set_index(['start'], inplace = True)
frequency
```

```python
#make 0-1 matrix
columns = list(start_end.columns)
mask = start_end.copy()
```

```python
for column in columns:
    mask[column] = mask[column].apply(lambda x: 1 if x>=5 else 0)
```

```python
mask
```

```python
def svt_solve(A, Omega, tau = None, delta = None, epsilon = 1e-2, max_iterations = 1500):
    Y = np.zeros_like(A)
    if not tau:
        tau = 5*np.sum(A.shape)/2
    if not delta:
        delta = 1.2*np.prod(A.shape)/np.sum(Omega)
    for _ in range(max_iterations):
        U, S, V = np.linalg.svd(Y, full_matrices = False)
        S = np.maximum(S-tau, 0)
        X = np.linalg.multi_dot([U, np.diag(S), V])
        Y += delta * Omega * (A - X)
        rel_recon_error = np.linalg.norm(Omega * (X - A) / np.linalg.norm(Omega * A))
        if rel_recon_error < epsilon:
            break
    return X
            
```

```python
data = np.array(frequency)
Omega = np.array(mask)
```

```python
X_filled_in = svt_solve(frequency, Omega)
```

```python
X_filled_in
```

```python
#用补全后的值来代替0值
for i in range(frequency.shape[0]):
    for j in range(frequency.shape[1]):
        if frequency.iloc[i,j] == 0:
            frequency.iloc[i,j] = X_filled_in[i,j]
```

```python
#用0进行截断
for i in range(frequency.shape[0]):
    for j in range(frequency.shape[1]):
        if frequency.iloc[i,j] < 0:
            frequency.iloc[i,j] = 0
```
